# Show at least 2 graphs about the data, it can be anything 

```python
import matplotlib.pyplot as plt
# in order to create bar chart
sum_y_by_x = accidents_subset.groupby('YEARMONTH')['DEAD'].sum()

# Plot the data
sum_y_by_x.plot(kind='bar')

# Add labels and title
plt.xlabel('YEARMONTH')
plt.ylabel('Sum of DEAD')
plt.title('Sum of DEAD for each YEARMONTH')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
sum_y_by_x = accidents_subset.groupby('CITY')['_id'].count()
top_10_sums = sum_y_by_x.nlargest(10)
# Plot the data
top_10_sums.plot(kind='bar')

# Add labels and title
plt.xlabel('ciry')
plt.ylabel('number of accidents')
plt.title('number of accidents per city')

# Show the plot
plt.show()
```

## Create at least 2 new columns in the Excel file, representing new features about the data.

```python
import numpy as np
percentiles_lives = accidents_subset['INJTOTAL'].transform(lambda x: pd.qcut(x, q=3, labels=['low', 'medium', 'high'], duplicates='drop'))
percentiles_money = accidents_subset['VEHICLE'].transform(lambda x: pd.qcut(x, q=3, labels=['low', 'medium', 'high'], duplicates='drop'))

conditions_lives = [
    (percentiles_lives == 'low'),
    (percentiles_lives == 'medium'),
    (percentiles_lives == 'high')
]
conditions_money = [
    (percentiles_money == 'low'),
    (percentiles_money == 'medium'),
    (percentiles_money == 'high')
]
values_damage_money = ['low', 'medium', 'high']
values_damage_lives = ['low', 'medium', 'deadly']

# Adding new columns based on conditions
accidents_subset['damage_money'] = np.select(conditions_lives, values_damage_money, default='unknown')
accidents_subset['damage_lives'] = np.select(conditions_money, values_damage_lives, default='N/A')

accidents_subset
```
